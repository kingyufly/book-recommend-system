```python
#!/usr/bin/env python
# coding: utf-8

from telepot.namedtuple import InlineKeyboardMarkup,InlineKeyboardButton
import time,requests,telepot,json
from telepot.loop import MessageLoop

answered_flag = False

def handle(msg):
    global answered_flag
    content_type, chat_type, chat_id = telepot.glance(msg)
    
    print(content_type)
    if content_type == "text":
        content = msg["text"]
        if (content == "/start"):
            # Send HTTP request
            response = requests.post("http://127.0.0.1:8080/register", data = json.dumps({"chat_id": str(chat_id)}))
            json_data = json.loads(response.text)
            result = json_data.get("exists")
            if result == 1:
                bot.sendMessage(chat_id, "Welcome back!")
            else:
                bot.sendMessage(chat_id, "Welcome!")
        elif (content == "/rate"):
            response = requests.post("http://127.0.0.1:8080/get_unrated_movie", data = json.dumps({"chat_id": str(chat_id)}))
            if(response.text == "{}"):
                bot.sendMessage(chat_id, "You have not registered yet! Please input /start first")
            else:
                json_data = json.loads(response.text)
                title = json_data.get("title")
                url = json_data.get("url")
                movie_id = json_data.get("movie_id")

                bot.sendMessage(chat_id, title + ":" + url)
                keyboard = InlineKeyboardMarkup(inline_keyboard=[
                    [InlineKeyboardButton(text='1', callback_data="{\"movie_id\":\"" + movie_id + "\",\"data\":\"1\"}"),
                     InlineKeyboardButton(text='2', callback_data="{\"movie_id\":\"" + movie_id + "\",\"data\":\"2\"}"),
                     InlineKeyboardButton(text='3', callback_data="{\"movie_id\":\"" + movie_id + "\",\"data\":\"3\"}"),
                     InlineKeyboardButton(text='4', callback_data="{\"movie_id\":\"" + movie_id + "\",\"data\":\"4\"}"),
                     InlineKeyboardButton(text='5', callback_data="{\"movie_id\":\"" + movie_id + "\",\"data\":\"5\"}")]
                ])
                bot.sendMessage(chat_id, 'How do you rate this movie?', reply_markup=keyboard)
                answered_flag = False
        elif (content == "/recommend"):
            bot.sendMessage(chat_id, "Please wait 10 or more seconds to get the recommended movies :)")
            response = requests.post("http://127.0.0.1:8080/recommend", data = json.dumps({"chat_id": str(chat_id),"top_n": 3}))
            if(response.text == "{}"):
                bot.sendMessage(chat_id, "You have not registered yet! Please input /start first")
            else:    
                json_data = json.loads(response.text)
                movies = json_data.get("movies")
                if(len(movies) == 0):
                    bot.sendMessage(chat_id, "You have not rated enough movies, we cannot generate recommendation for you")
                else:
                    bot.sendMessage(chat_id, "Here are the recommended movies:")
                    for movie in movies:
                        title = movie["title"]
                        url = movie["url"]
                        bot.sendMessage(chat_id, title + ":" + url)
        elif content[0] == '/':
            bot.sendMessage(chat_id, "Unknown Command! Please try again")
        else:
            bot.sendMessage(chat_id, "Unknown Command! Please input the command start with '/'")

def on_callback_query(msg):
    # the rate can only be answered once
    global answered_flag
    query_id, from_id, query_data = telepot.glance(msg, flavor='callback_query')
    if(answered_flag == True):
        bot.answerCallbackQuery(query_id, text='You have rate this movie! For rating more movie, please type command /rate')
    else:
        json_data = json.loads(query_data)

        bot.answerCallbackQuery(query_id, text='Thanks for your rating!')
        answered_flag = True
        response = requests.post("http://127.0.0.1:8080/rate_movie",
                                 data = json.dumps(
                                     {"chat_id": from_id,
                                      "movie_id": json_data.get("movie_id"),
                                      "rating": json_data.get("data"),
                                     }
                                 )
                                )
if __name__ == "__main__":
    # Provide your bot's token
    bot = telepot.Bot("787863087:AAE8fUa16n9lMxMTFNEwuN6zoYJORlC7MTE")
    MessageLoop(bot, {'chat': handle, 'callback_query': on_callback_query}).run_as_thread()
    while True:
        time.sleep(10)
```
